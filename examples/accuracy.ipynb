```python
import buildtools as bt

import numpy as np
import matplotlib.pyplot as plt
import pprint
```

```python
# Case 1: Randomly Paired Data
```

```python
n_xy = 10
n_ab = 10

xy,ab = np.random.rand(n_xy,2), np.random.rand(n_ab,2)
bt.plot_pairs(xy,ab)

# This calculates all relevant statistics; t is a distance threshold
result = bt.build_score(xy,ab,t=.1)
pprint.pprint(result)
```

# Case 2: Equal Numbers of XY (true) and AB (detected)

```python
n_xy = 10
n_ab = 10

xy,ab = np.random.rand(n_xy,2), np.random.rand(n_ab,2)
result = bt.build_score(xy,ab,t=.1)
pprint.pprint(result)

bt.plot_pairs(xy,ab,result['hungarian_indexes'])
```

# Case 3: More observations in ab (simulating noise)
